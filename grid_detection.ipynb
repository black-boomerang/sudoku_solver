```python
import warnings

import cv2 as cv

from grid_detection import LineType, detect_grid
from utils import show_image

warnings.filterwarnings('ignore')
```

```python
image = cv.imread('sudoku_dataset/images/image10.jpg')
h_lines, v_lines = detect_grid(image, show_steps=True)
```

```python
image_copy = image.copy()

for line in v_lines + h_lines:
    x1, y1, x2, y2 = line.x1, line.y1, line.x2, line.y2
    if line.line_type == LineType.HORIZONTAL:
        if line.x1 > line.x2:
            x1 = line.x1 + 1000
            x2 = line.x2 - 1000
        else:
            x1 = line.x1 - 1000
            x2 = line.x2 + 1000
    else:
        if line.y1 > line.y2:
            y1 = line.y1 + 1000
            y2 = line.y2 - 1000
        else:
            y1 = line.y1 - 1000
            y2 = line.y2 + 1000

    cv.line(image_copy, (x1, y1), (x2, y2), (255, 0, 0), 2)

show_image(image_copy, 'Результат детекции')
```
